```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

Convert Dataset to Long Format

```python
df_long = df.melt(id_vars=["Country Name"],
                  var_name="Year",
                  value_name="CO2_Emission")

df_long.head()
```

Convert Year to Integer

```python
df_long["Year"] = df_long["Year"].astype(int)
```

Remove Missing Values

```python
df_long = df_long.dropna()
```

Global Emission Trend

```python
global_trend = df_long.groupby("Year")["CO2_Emission"].mean()

plt.figure(figsize=(12,6))
plt.plot(global_trend)
plt.title("Global CO2 Emission Trend (1960-2018)")
plt.xlabel("Year")
plt.ylabel("Average CO2 Emission")
plt.show()
```

Top 10 Highest Emitting Countries

```python
top_countries = df_long.groupby("Country Name")["CO2_Emission"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
top_countries.plot(kind="bar")
plt.title("Top 10 CO2 Emitting Countries")
plt.ylabel("Average CO2 Emission")
plt.show()
```

CO2 Emissions Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df_long["CO2_Emission"], bins=30)
plt.title("Distribution of CO2 Emissions")
plt.show()
```

Country Trend Example

```python
india = df_long[df_long["Country Name"]=="India"]

plt.figure(figsize=(10,5))
plt.plot(india["Year"], india["CO2_Emission"])
plt.title("CO2 Emissions Trend - India")
plt.xlabel("Year")
plt.ylabel("CO2 Emission")
plt.show()
```

Correlation Heatmap (Recent Years)

```python
recent = df.iloc[:, -10:]  # last 10 years

plt.figure(figsize=(10,6))
sns.heatmap(recent.corr(), cmap="coolwarm")
plt.title("Correlation Between Recent Years CO2 Emissions")
plt.show()
```

ALGORITHM USED

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Step 2: Load sensor data
data = pd.read_csv("co2_emissions.csv")

# Step 3: Preprocess
X = data.drop("CO2_emission", axis=1)  # Features
y = data["CO2_emission"]               # Target

# Step 4: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Step 5: Model training
model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, y_train)

# Step 6: Prediction & evaluation
y_pred = model.predict(X_test)
print("MSE:", mean_squared_error(y_test, y_pred))
```
